```python
#NLTK Stemming
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
words = ["eating", "eats", "eat", "ate", "adjustable", "rafting", "ability", "meeting"]

for word in words:
    print(word, " | ", stemmer.stem(word))
```

```python
#SPACY Stemming

import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("eating eats eat ate adjustable rafting ability meeting better")

for token in doc:
    print(token," | ", token.lemma_)

doc = nlp("I jogged for three hours although I am not casual in it.")
for token in doc:
    print(token," | ", token.lemma_)
```

```python
nlp.pipe_names
```

```python
ar = nlp.get_pipe('attribute_ruler')
ar.add([[{"TEXT":"Yeah"}],[{"TEXT": "yupe"}],[{"TEXT":"yo"}]],{"LEMMA": "Yes"})
text = "Yeah I will be there and yupe from my side yo"
doc = nlp(text)

for token in doc:
    print(token, " | ", token.lemma_)
```
