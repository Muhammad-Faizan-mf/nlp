```python
#spacy
import spacy

nlp =  spacy.load("en_core_web_sm")
doc = nlp("My name is M.Faizan. I am doing my Masters in Natural Language Processing from University of Trier.")

for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```
