```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
with open('text.txt') as f:
    news = f.read()
news
```

```python
doc = nlp(news)
nouns = []
numerals = []

for token in doc:
    if token.pos_ == "NOUN":
        nouns.append(token)
    if token.pos_== "NUM":
        numerals.append(token)

print(f"Nouns : {nouns}")
print(f"Numeric : {numerals}")
```

```python
count = doc.count_by(spacy.attrs.POS)
count
```

```python
for k, v in count.items():
    print(doc.vocab[k].text, "=>", v)
```
