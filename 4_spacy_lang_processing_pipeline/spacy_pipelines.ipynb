```python
import spacy

nlp = spacy.blank('en')
doc  = nlp ("I ate pasta for $6. Then I want to have some drink.")

for token in doc:
    print(token)
```

```python
nlp.pipe_names
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.pipe_names
```

```python
nlp.pipeline
```

```python
doc = nlp("I ate pasta for $6. Then I want to have some drink.")

for token in doc:
    print(token, " | ", spacy.explain(token.pos_)," | ", token.lemma_)
```

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:
    print(ent.text , ent.label_)
```

```python
from spacy import displacy

displacy.render(doc, style="ent")
```
