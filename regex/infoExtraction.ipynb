```python
def match_pattern(pattern, text):
    matches = re.findall(pattern, text)
    if matches:
        return matches[0]
    
print(match_pattern(pattern,"The order 4312423432 was recieved late!"))
```

```python
email_pattern = r"[a-zA-Z0-1]*@[a-z]*\.[a-zA-Z0-9]*"
print(match_pattern(email_pattern,chat2))
```

```python
text='''
Born	Elon Reeve Musk
June 28, 1971 (age 50)
Pretoria, Transvaal, South Africa
Citizenship	
South Africa (1971–present)
Canada (1971–present)
United States (2002–present)
Education	University of Pennsylvania (BS, BA)
Title	
Founder, CEO and Chief Engineer of SpaceX
CEO and product architect of Tesla, Inc.
Founder of The Boring Company and X.com (now part of PayPal)
Co-founder of Neuralink, OpenAI, and Zip2
Spouse(s)	
Justine Wilson
​
​(m. 2000; div. 2008)​
Talulah Riley
​
​(m. 2010; div. 2012)​
​
​(m. 2013; div. 2016)
'''


def extract_personal_information(text):
    age = match_pattern(r"age (\d+)",text)
    name  = match_pattern(r"Born(.*)\n",text)
    birth_date = match_pattern(r"Born.*\n(.*)\(age",text)
    birth_place = match_pattern(r"\(age.*\n(.*)",text)
    citizen_of = match_pattern(r"Citizenship.*\n(.*)\(1971",text)
    education = match_pattern(r"Education(.*)",text)
    founder = match_pattern(r"Founder.*Company and(.*)\(now",text)
    co_founder = match_pattern(r"Co-founder of(.*)",text)

    return {
        'age': int(age),
        'name': name.strip(),
        'birth_date': birth_date.strip(),
        'birth_place': birth_place.strip(),
        'citizen_of': citizen_of.strip(),
        'founder' : founder.strip(),
        'education': education.strip(),
        'co_founder': co_founder.strip(),

    }
extract_personal_information(text)
```
